## **Conjuntos de Datos**

### **Juegos**

```python
juegos = pd.read_csv(ruta + 'juegos.csv')
juegos.info()
```

Se renombra la columna 'Unnamed: 0' como 'indice'.

```python
juegos.rename(columns={'Unnamed: 0':'juego_id'}, inplace=True)
juegos.info()
```

```python
juegos.head(5)
```

### **Deportes**

```python
deportes = pd.read_csv(ruta + 'deporte.csv')
deportes.info()
```

### **Deportistas**

```python
deportistas1 = pd.read_csv(ruta + 'deportista.csv')
deportistas1.info()
```

```python
deportistas1.shape
```

```python
deportistas2 = pd.read_csv(ruta + 'deportista2.csv')
deportistas2.info()
```

```python
deportistas2.shape
```

```python
deportistas = pd.concat([deportistas1, deportistas2], axis=0)
deportistas.info()
```

```python
deportistas.shape
```

```python
deportistas.head(5)
```

```python
deportistas_error = pd.read_csv(ruta + 'deportistaError.csv')
deportistas_error.info()
```

### **Países**

```python
paises = pd.read_csv(ruta + 'paises.csv')
paises.info()
```

```python
paises.rename(columns={'id':'pais_id'}, inplace=True)
paises.info()
```

```python
paises.head(5)
```

### **Deportistas - Países**

```python
equiposOlimpicos = paises.join(deportistas.set_index('equipo_id'), on='pais_id')
equiposOlimpicos.info()
```

```python
equiposOlimpicos.head(10)
```

### **Resultados**

```python
resultados = pd.read_csv(ruta + 'resultados.csv')
resultados.info()
```

```python
resultados.head(5)
```

```python
deportistas.head(5)
```

```python
resultados1 = pd.merge(resultados, deportistas, on='deportista_id', how='inner')
resultados1.info()
```

```python
resultados1.head(5)
```

```python
resultados2 = resultados.join(deportistas.set_index('deportista_id'), on='deportista_id')
resultados2.info()
```

```python
resultados2.head(5)
```

```python
juegos.info()
```

```python
resultados3 = resultados2.join(juegos.set_index('juego_id'), on='juego_id')
resultados3.info()
```

```python
resultados3.head(5)
```
