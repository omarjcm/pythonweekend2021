# **Olimpiadas**

A continuación, la carga de los datos con la inicialización de una sesión en Apache Spark.

```python
from pyspark import SparkContext
sc = SparkContext(master = "local", appName = "Olimpiadas-RDD") 

ruta = '/content/drive/MyDrive/CIENCIA DE DATOS/PythonWeekend_2021/datos/olimpiadas/'
```

## **Conjuntos de Datos**

### **Países**

Los países serían los equipos olímpicos.

```python
equiposOlimpicosRDD = sc.textFile(ruta+"paises.csv").map(lambda line : line.split(","))
equiposOlimpicosRDD.take(10)
```

### **Deportistas**

```python
deportistaOlimpico1RDD = sc.textFile(ruta+"deportista.csv").map(lambda line : line.split(","))
deportistaOlimpico2RDD = sc.textFile(ruta+"deportista2.csv").map(lambda line : line.split(","))
deportistaOlimpicoRDD = deportistaOlimpico1RDD.union(deportistaOlimpico2RDD)
```

```python
deportistaOlimpicoRDD.take(5)
```

Con la función takeSample, puedo obtener una cantidad de datos aleatoriamente seleccionada, los argumentos son:


*   num = número de elementos a mostrar
*   withReplacement = si la muestra tendrá valores repetidos o no.
*   sed = la semilla para hacer reproducible la toma de valores aleatorios.



```python
deportistaOlimpicoRDD.takeSample(num=10, withReplacement=False, seed=123)
```

### **Deportistas - Países**

```python
equiposOlimpicosRDD.map(lambda x : [x[0],x[2]]).take(10)
```

```python
deportistaOlimpicoRDD.map(lambda x :[x[-1], x[:-1]]).take(10)
```

```python
equipoDeportista = deportistaOlimpicoRDD.map(lambda x :[x[-1], x[:-1]]).join(equiposOlimpicosRDD \
                        .map(lambda x : [x[0],x[2]] ))
```

```python
equipoDeportista.take(10)
```

### **Resultados**

```python
resultado = sc.textFile(ruta+"resultados.csv").map(lambda line : line.split(","))
resultado.take(5)
```

### **Deportista - País - Medalla**

Se eliminan las medallas no ganadoras (medalla = 'NA') y se quedan todas las medallas que tengan valores.

```python
resultadoGanador = resultado.filter(lambda l : 'NA' not in l[1])
resultadoGanador.take(5)
```

```python
resultadoGanador = resultadoGanador.map(lambda l : [l[2],l[1]])
resultadoGanador.take(5)
```

```python
equipoDeportista.take(5)
```

```python
resultadoGanador.take(5)
```

```python
jugadoresMedalla =  equipoDeportista.join(resultadoGanador)
jugadoresMedalla.take(10)
```

### [4].- Contar cuántos registros tiene un conjunto de datos.

Se lo puede realizar de dos maneras, cuando son pocos directamente como se encuentra a continuación:

```python
deportistaOlimpicoRDD.map(lambda x :[x[-1], x[:-1]]).join(equiposOlimpicosRDD \
                            .map(lambda x : [x[0],x[2]])).count()
```

Sin embargo, cuando son muchos registros, lo mejor es utilizar la función countApprox y colocar los milisegundos que se tarde en contar la cantidad de registros, esta función no me daría una cantidad exacta pero si aproximada.

```python
deportistaOlimpicoRDD.map(lambda x :[x[-1], x[:-1]]).join(equiposOlimpicosRDD \
                        .map(lambda x : [x[0],x[2]])).countApprox(20)
```

En este caso, sale el mismo valor con **count** y **countApprox** debido a que no son muchos registros a contabilizar.

### [5].- Obtener los puntajes históricos de los países jugadores.

```python
jugadoresMedalla.take(10)
```

```python
puntajes = {'Gold':7, 'Silver':5, 'Bronze':4}
paisesMedallas = jugadoresMedalla.map(lambda x : (x[1][0][-1], puntajes[x[1][1]]) )
paisesMedallas.takeSample(False,10)
```

### [6].- Valores puntajes históricos de los países jugadores.

```python
from operator import add
conclusion = paisesMedallas.reduceByKey((add)).sortBy(lambda x : x[1], ascending=False)
conclusion.take(10)
```
