# **Olimpiadas**

```python
import pyspark
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName('Olimpiadas-Dataframe').getOrCreate()

ruta = '/content/drive/MyDrive/CIENCIA DE DATOS/PythonWeekend_2021/datos/olimpiadas/'
```

## **Conjuntos de Datos**

### **Países**

```python
equiposOlimpicos = spark.read.format('csv').option('inferSchema', True).option('header', True).load(ruta + 'paises.csv')
equiposOlimpicos.show()
```

### **Deportistas**

```python
deportistas1 = spark.read.format('csv').option('inferSchema', True).option('header', True).load(ruta + 'deportista.csv')
deportistas2 = spark.read.format('csv').option('inferSchema', True).option('header', True).load(ruta + 'deportista2.csv')

deportistas1.show()
```

```python
estado_deportistas1 = (deportistas1.count(), len(deportistas1.columns))
estado_deportistas1
```

```python
deportistas2.show()
```

```python
estado_deportistas2 = (deportistas2.count(), len(deportistas2.columns))
estado_deportistas2
```

```python
deportistas = deportistas1.union(deportistas2)
deportistas.show()
```

```python
estado_deportistas = (deportistas.count(), len(deportistas.columns))
estado_deportistas
```

### **Deportistas - Países**

```python
deportistas.printSchema()
```

```python
equiposOlimpicos.printSchema()
```

```python
equipoDeportista = deportistas.join(equiposOlimpicos, deportistas.equipo_id == equiposOlimpicos.id)
equipoDeportista.printSchema()
```

```python
equipoDeportista.show()
```

### **Resultados**

```python
resultados = spark.read.format('csv').option('inferSchema', True).option('header', True).load(ruta + 'resultados.csv')
resultados.show()
```

### **Deportista - País - Medalla**

```python
resultadoGanador = resultados.select('*').filter(resultados.medalla != 'NA')
resultadoGanador.show()
```

```python
resultadoGanador = resultadoGanador.select('deportista_id', 'medalla')
resultadoGanador.show()
```

```python
equipoDeportista.printSchema()
```

```python
resultadoGanador.printSchema()
```

```python
jugadoresMedalla = resultadoGanador.join(equipoDeportista, how='inner', on=['deportista_id'])
jugadoresMedalla.printSchema()
```

## **Consultas**

```python
from pyspark.sql.functions import count, desc, col, max, struct, approx_count_distinct
```

### [1].- ¿Cuántos equipos olímpicos existen?

```python
q1 = equiposOlimpicos.select('sigla').distinct().agg( count('*').alias('count') )
q1.show()
```

### [2].- ¿Cuántos equipos posee cada país?

```python
q2 = equiposOlimpicos.select('sigla').groupBy('sigla').agg( count('*').alias('count') )
q2.show()
```

```python
q3 = equiposOlimpicos.select('sigla', 'equipo') \
        .groupBy('sigla', 'equipo') \
        .agg( struct(col('equipo')) ) \
        .select('sigla', 'equipo') \
        .orderBy('sigla')
q3.show()
```

### [3].- Listar los equipos de Argentina

```python
q4 = equiposOlimpicos.filter(equiposOlimpicos.sigla == 'ARG')
q4.show()
```
